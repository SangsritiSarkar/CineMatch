### Transform genres column

```python
df.iloc[0].genres
```

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
convert(df.iloc[0].genres)
```

```python
df['genres'] = df['genres'].apply(convert)
```

### Transform keywords column


```python
df['keywords'] = df['keywords'].apply(convert)
```

```python
df.head()
```

### Transform cast column

```python
df['cast'][0]
```

```python
def convert2(obj):
    L = []
    cnt = 0
    for i in ast.literal_eval(obj):
        if cnt!=3:
            L.append(i['character'])
            cnt+=1
        else:
            break
    return L
```

```python
df['cast'] = df['cast'].apply(convert2)
```

```python
df.head()
```

### Transform crew column

```python
def fetch_director(obj):
    L = []
    cnt = 0
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
    return L
```

```python
df['crew'] = df['crew'].apply(fetch_director)
```

```python
df.head()
```

### Transform overview column from string to list

```python
df['overview'] = df['overview'].apply(lambda x:x.split())
```

```python
df.head()
```

### Removing spaces in-between elements

```python
df['genres'] = df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
df['keywords'] = df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
df['cast'] = df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df['crew'] = df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df.head()
```

### Concatenate all tag columns

```python
df['tags'] = df['overview']+df['genres']+df['keywords']+df['cast']+df['crew']
```

```python
df.head()
```

```python
new_df = df[['movie_id','title','tags']]
```

### Convert tags column from list to string separated with spaces

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

### Transform each letter in tags to lower case

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

## Vectorization


### Removing similar words using NLP library nltk

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

### Vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
vectors.shape
```

```python
print(cv.get_feature_names_out())
```

```python
len(cv.get_feature_names_out())
```

### Calculate Similarities

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[0]
```

```python
similarity[0].shape
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
